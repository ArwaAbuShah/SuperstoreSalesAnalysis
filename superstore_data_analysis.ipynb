```python
df = pd.read_excel('Sample - Superstore.xls')
```

## Initial Data Exploration

```python
print("First 5 rows of the dataset:")
print(df.head())
```

```python
print("Dataset Info (columns, data types, non-null counts):")
print(df.info())
```

```python
print("Dataset Shape (rows, columns):")
print(df.shape)
```

## Basic Data Quality Checks

```python
print("Checking for missing values in each column:")
print(df.isnull().sum())
```

```python
print("Checking for duplicate rows:")
print(f"Number of perfect duplicate rows: {df.duplicated().sum()}")
```

## Simple Analysis & Visualization

```python
# The total sales and profit by category
print("Total Sales and Profit by Category:")
category_summary = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values(by='Profit', ascending=False)
print(category_summary)
```

```python
# A bar chart for Profit by Category
plt.figure(figsize=(10, 5))
category_summary['Profit'].plot(kind='bar', color='skyblue')
plt.title('Total Profit by Product Category')
plt.xlabel('Category')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# A line chart for Sales over time 
df['Order Date'] = pd.to_datetime(df['Order Date']) # Convert to datetime
sales_over_time = df.groupby('Order Date')['Sales'].sum()

plt.figure(figsize=(12, 5))
sales_over_time.plot(kind='line', color='green')
plt.title('Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales ($)')
plt.show()
```
